```
import zipfile
zip_ref = zipfile.ZipFile('/content/train.zip', 'r')
zip_ref.extractall('/content')
zip_ref.close()
```

```
from tensorflow.keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
```

```
img=image.load_img('/content/train/laminar.png',target_size=(200,200))
```

```
import matplotlib.pyplot as plt
```

```
plt.imshow(img)
```

```
datagen=ImageDataGenerator(
    rotation_range=30,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2,
)
```

```
img=image.img_to_array(img)
```

```
img.shape
```

```
input_batch=img.reshape(1,200,200,3)
```

```
zip_ref = zipfile.ZipFile('/aug.zip', 'r')
zip_ref.extractall('/content')
zip_ref.close()
```

```
i=0
for output in datagen.flow(input_batch,batch_size=1,save_to_dir='train'):
  i=i+1
  if i==10:
    break
```
